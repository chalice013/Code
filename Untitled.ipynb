# 解壓縮 & 合併檔案

```python
# coding=UTF-8

import os , sys, tarfile
from datetime import datetime

class Unzip_Merge_All():
    
    def getzipfiles(self,file_dir):
        datetimelist = os.listdir(file_dir)
        return datetimelist

    def unzip(self,url,path):
        try:
            tar = tarfile.open(url, 'r:gz')
            tar.extractall(path = path , members = None)
        except:
            name = os.path.basename(sys.argv[0])
            print(name[:name.rfind('.')], 'not find ')
    
    def getcsvfiles(self,path): 
        ls = []
        for dirPath, dirNames, fileNames in os.walk(path):
            for f in fileNames:
                ls.append(os.path.join(dirPath, f))
        return ls
   
    def mergefromlist(self , file_list , path):
        for filename in file_list:
            mergefilename = path + filename.split('_')[-2] + '.csv'
            with open(mergefilename , 'a') as f:
                with open(filename , 'r') as f2:
                    f.write(f2.read())
        
if __name__ == '__main__':

    file_dir = raw_input('tar.gz files come from : ') # e:\etc\try\
    unzip_path = raw_input('Unzip files store to : ') # e:\etc\try2\

    e = Unzip_Merge_All()
    
    for ent in e.getzipfiles(file_dir):
        e.unzip(file_dir + ent , unzip_path)
    
    print 'Unzip Done' 
    
    csvfile_path = raw_input('Merged csv files store to : ') # 'e:\etc\try3\'

    for ent in os.listdir(unzip_path):
        path = unzip_path + ent + '/'
        #print path2 
        ls = e.getcsvfiles(path)
        e.mergefromlist(ls, csvfile_path)

    print 'Merge Done'
```
